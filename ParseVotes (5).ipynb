```
without_votes = []
def get_groups(vote_id,t):
  gr_list = []
  flag = -1
  try:
    for i in raw_vote_data_in_json[vote_id]['votes'][t]['groups']: 
      #print(i)     
      gr_list.append(i)
  except KeyError:
    flag = 1
  if (flag > -1):
    without_votes.append(vote_id) #  надо добавить удаление дупликатов!
  #cache_votes = list(dict.fromkeys(without_votes))
  #without_votes = cache_votes
  res_list=list(dict.fromkeys(gr_list))
  return res_list
```

```
err_votes_nums=[]
```

```
def get_num(vote_num): #сейчас она возвращает чисто лист депутатов для конкретного голосования без деления по голосам
  meps = []
  ter = get_groups(vote_num,'+')
  for gr in ter:
    for i in range(len(raw_vote_data_in_json[vote_num]['votes']['+']['groups'][gr][:])):
      try:
          meps.append(raw_vote_data_in_json[vote_num]['votes']['+']['groups'][gr][i]['mepid'])
      except KeyError:
          srtyy = 0
          err_votes_nums.append("+__"+str(i))
  ter = get_groups(vote_num,'-')
  for gr in ter:
    for i in range(len(raw_vote_data_in_json[vote_num]['votes']['-']['groups'][gr][:])):
      try:
          meps.append(raw_vote_data_in_json[vote_num]['votes']['-']['groups'][gr][i]['mepid'])
      except KeyError:
          srtyy = 0
          err_votes_nums.append("-__"+str(i))
  ter = get_groups(vote_num,'0')
  for gr in ter:
    for i in range(len(raw_vote_data_in_json[vote_num]['votes']['0']['groups'][gr][:])):
      try:
          meps.append(raw_vote_data_in_json[vote_num]['votes']['0']['groups'][gr][i]['mepid'])
      except KeyError:
          srtyy = 0
          err_votes_nums.append("0__"+str(i))   
  return meps
```

```
def get_vote(mep_num,vote_num): #сейчас она возвращает чисто лист депутатов для конкретного голосования без деления по голосам
  res_data = 0
  ter = get_groups(vote_num,'+')
  for gr in ter:
    for i in range(len(raw_vote_data_in_json[vote_num]['votes']['+']['groups'][gr][:])):
      try:
        if (raw_vote_data_in_json[vote_num]['votes']['+']['groups'][gr][i]['mepid'] == mep_num):
          res_data = +1
      except KeyError:
        res_data=0
  ter = get_groups(vote_num,'-')
  for gr in ter:
    for i in range(len(raw_vote_data_in_json[vote_num]['votes']['-']['groups'][gr][:])):
      try:
        if (raw_vote_data_in_json[vote_num]['votes']['-']['groups'][gr][i]['mepid'] == mep_num):
          res_data = -1  
      except KeyError:
        res_data=0
  return res_data
```
